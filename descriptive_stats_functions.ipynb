```python
''' Reads a text file and calculates the mean, median, mode,
variance, and standard deviation. '''

''' *** Variance and standard deviation are being calulated for the population and not a sample *** '''

# function to read in text file
def read_data(txt_file):
    numbers = []
    with open(txt_file) as f:
        for line in f:
            numbers.append(float(line))
            
    return numbers

# function to calculate mean
def calculate_mean(numbers):
    s = sum(numbers)
    N = len(numbers)
    mean = s/N
    
    return mean

# function to calculate median
def calculate_median(numbers):
    N = len(numbers)
    numbers.sort()

    # Find the median
    if N % 2 == 0:
        # if N is even
        m1 = N/2
        m2 = (N/2) + 1
        # Convert to integer, match position
        m1 = int(m1) - 1
        m2 = int(m2) - 1
        median = (numbers[m1] + numbers[m2])/2
    else:
        m = (N+1)/2
        # Convert to integer, match position
        m = int(m) - 1
        median = numbers[m]
        
    return median

# function to calculate mode
from collections import Counter
def calculate_mode(numbers):
    c = Counter(numbers)
    numbers_freq = c.most_common()
    max_count = numbers_freq[0][1]
    modes = []
    for num in numbers_freq:
        if num[1] == max_count:
            modes.append(num[0])
            
    str_mode = 'Mode: '
    for index, mode in enumerate(modes):
        str_mode += str(mode)
        
        if index < len(modes)-1:
            str_mode +=  ', '
            
    return str_mode

# function to calculate variance
def find_differences(numbers):
    # Find the mean
    mean = calculate_mean(numbers)
    # Find the differences from the mean
    diff = []
    for num in numbers:
        diff.append(num-mean)
    return diff


def calculate_variance(numbers):
    # Find the list of differences
    diff = find_differences(numbers)
    # Find the squared differences
    squared_diff = []
    
    for d in diff:
        squared_diff.append(d**2)
    # Find the variance
    sum_squared_diff = sum(squared_diff)
    variance = sum_squared_diff/len(numbers)
    
    return variance

if __name__ == '__main__':
    txt_file_data = read_data('my_data.txt')
    
    # calculate mean
    mean = calculate_mean(txt_file_data)
    print(f'Mean: {mean}')
    
    # calculate median
    median = calculate_median(txt_file_data)
    print(f'Median: {median}')

    # calculate mode
    str_mode = calculate_mode(txt_file_data)
    print(str_mode)
    
    # calculate variance
    variance = calculate_variance(txt_file_data)
    print(f'Variance: {variance}' )

    # calculate standard deviation
    std_dev = variance**0.5
    print(f'Standard Deviation: {std_dev}')
```
